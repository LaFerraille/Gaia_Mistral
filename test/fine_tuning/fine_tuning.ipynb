```python
import torch
import transformers
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model, PeftModel
```

## Loading Base Model

```python
tokenizer = AutoTokenizer.from_pretrained("mistralai/Mixtral-8x7B-Instruct-v0.1")
```

```python
model = AutoModelForCausalLM.from_pretrained("mistralai/Mixtral-8x7B-Instruct-v0.1",
                                             load_in_4bit=True,
                                             torch_dtype=torch.float16,
                                             device_map="auto")                                             )
```

## LoRA Model Preparation

```python
tokenizer.pad_token = '!'
```

```python
CUTOFF_LEN = 256
LORA_R = 8
LORA_ALPHA = 2 * LORA_R
LORA_DROPOUT = 0.1
```

```python
config = LoraConfig(
    r=LORA_R,
    lora_alpha=LORA_ALPHA,
    target_modules=[ "w1", "w2", "w3"],
    lora_dropout=LORA_DROPOUT,
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
model = get_peft_model(model, config)
```

## Loading and Pre-Processing Data

```python
df = pd.read_csv("data.csv")
df.drop(columns=["url"],inplace=True)
df= df.dropna()
```

```python
df[df['title'].isnull() | df['description'].isnull()]
```

```python
dataset = Dataset.from_pandas(df)
```

```python
dataset
```

```python
train_data = dataset
```

```python
def generate_prompt(user_query,  sep="\n\n### "):
    sys_msg= "Vous êtes un assistant pour les agriculteurs de France"
    p =  "<s> [INST]" + sys_msg +"\n"+ user_query["title"] + "[/INST]" +  user_query["description"] + "</s>"
    return p
```

```python
def tokenize(prompt):
    return tokenizer(
        prompt + tokenizer.eos_token,
        truncation=True,
        max_length=CUTOFF_LEN ,
        padding="max_length"
    )
```

```python
train_data = train_data.shuffle()
train_data[0]
```

```python
train_data = train_data.map(lambda x: tokenize(generate_prompt(x)))
train_data[0]
```

```python
train_data = train_data = train_data.remove_columns(['title', 'description', '__index_level_0__'])
train_data[0]
```

## Training

```python
trainer = Trainer(
    model=model,
    train_dataset=train_data,
    args=TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        num_train_epochs=6,
        learning_rate=1e-4,
        logging_steps=2,
        optim="adamw_torch",
        save_strategy="epoch",
        output_dir="mixtral-fine-tuned"
    ),
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False)
)
model.config.use_cache = False
```

```python
trainer.train()
```

## Doing inferences

```python
def predict(input_text, model, tokenizer, max_length=256):
    inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=max_length, truncation=True)
    device = model.device
    inputs = inputs.to(device)
    with torch.no_grad():
        outputs = model.generate(inputs, max_length=max_length)
    predicted_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return predicted_text
```

```python
model.eval()
```

```python
input_text = "comment déclarer l'achat de mon champ?"
predicted_text = predict(input_text, model, tokenizer)
print(predicted_text)
```
